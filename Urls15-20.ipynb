```python
import json, urllib.request
import pandas as pd

f = open("urls15.txt", "r")

for url in f.readlines():
    with urllib.request.urlopen(url) as response:
        print('processing: ', url)
        resp = json.loads(response.read())
        data = resp.get('data')
        if data is not None:
            print('found data for: ', url)
            df = pd.DataFrame(data)
            df.to_csv('Assam_2015-2020.csv', mode='a', index = False, header=None)
```
